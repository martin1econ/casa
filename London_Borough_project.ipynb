```python
def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['Year']==1998])
    y2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
def hratio(filt, beg_year, end_year):
    """return ratio based on conditional and two dates(years)"""
    filt_yr1= (dfg['Year']==beg_year)
    filt_yr2= (dfg['Year']==end_year)
    beg = float(dfg.loc[filt_yr1 & filt,'Average_price'])
    end = float(dfg.loc[filt_yr2 & filt,'Average_price'])
    ratio = end/beg
    return ratio
    
    
```

```python
hratio(dfg['London_Borough']=='Barking & Dagenham',1998,2018)
```

```python
final={}
```

```python
for b in dfg['London_Borough'].unique():
    borough = dfg['London_Borough']==b
    final[b] = hratio(borough,1998,2018)
    
```
